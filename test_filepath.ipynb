```python
import sqlite3
import pandas as pd
import os # For 'chat.db' file location lookup
```

```python
# Lookup the full filepath to 'chat.db'
file = os.path.abspath("chat.db")

# Create a read-only connection to the database
connection = sqlite3.connect(file, uri=True)

# Show the tables of the database
tables = pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", connection)
display(tables)

# Select all columns and rows from chat table
df = pd.read_sql_query("SELECT * FROM chat", connection)
# Change settings to allow output of all column names
pd.options.display.max_columns = None
pd.options.display.max_rows = None

# Display the dataframe
display(df)

# Close the connection
connection.commit()
connection.close()
```
